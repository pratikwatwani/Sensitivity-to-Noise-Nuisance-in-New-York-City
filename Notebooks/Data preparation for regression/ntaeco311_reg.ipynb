```python
import pandas as pd
import numpy as np

df = pd.read_csv('nta_eco311.csv')
df.head()
```

```python
df.info(verbose=True, null_counts=True)
```

```python
cols = ['Unnamed: 0','Unnamed: 0.1','created_date','closed_date','agency','agency_name','complaint_type','descriptor','location_type','street_name','cross_street_1','cross_street_2','intersection_street_1','intersection_street_2','address_type','city','landmark','facility_type','status','due_date','resolution_description','resolution_action_updated_date']
```

```python
temp_df = df.copy()
temp_df.drop(cols, axis = 1, inplace = True)
temp_df.head()
```

```python
df_pivot = pd.pivot_table(df, index = df.NTA_code)
```

```python
crime_total = temp_df.NTA_code.value_counts()
crime_total = pd.DataFrame(crime_total)
crime_total.columns = ['Counts']
```

```python
df_pivot.columns = df_pivot.columns.str.strip()
```

```python
df_pivot.columns
```

```python
df_pivot_insur = df_pivot.loc[:,df_pivot.columns[:7]]
```

```python
df_pivot.drop(df_pivot.columns[:7], axis = 1, inplace = True)
```

```python
df_pivot_salary = df_pivot.loc[:,df_pivot.columns[2:15]]
df_pivot.drop(df_pivot.columns[2:15], axis = 1, inplace = True)
```

```python
df_pivot.drop(['Unnamed: 0', 'Unnamed: 0.1','index_right'], axis=1, inplace= True)
```

```python
df_pivot_jobs = df_pivot.loc[:,:]
```

```python
df_pivot_insur['X_insur'] = df_pivot_insur.sum(axis = 1)
df_pivot_salary['X_salary'] = df_pivot_salary.sum(axis = 1)
df_pivot_jobs['X_insur'] = df_pivot_jobs.sum(axis = 1)


df_pivot_insur = df_pivot_insur.merge(crime_total, on = df_pivot_insur.index)
df_pivot_salary = df_pivot_salary.merge(crime_total, on = df_pivot_salary.index)
df_pivot_jobs = df_pivot_jobs.merge(crime_total, on = df_pivot_jobs.index)
```

```python
df_pivot_salary.to_csv('DemoBySalary.csv')
df_pivot_insur.to_csv('DemoByInsurance.csv')
df_pivot_jobs.to_csv('DemoByJobs.csv')
```
