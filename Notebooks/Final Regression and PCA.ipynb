```python
sto = df1.merge(df2, left_on = 'Job0', right_on = 'Race0')
sto = sto.merge(df3,left_on = 'Job0', right_on = 'Edu0' )
sto = sto.merge(df6,left_on = 'Job0', right_on = 'Or0' )
sto = sto.merge(df7,left_on = 'Job0', right_on = 'Sal0' )
```

```python
sto.head()
```

```python
sto = sto.merge(total_populations, left_on = 'Job0', right_on = total_populations.index)
```

```python
sto.columns
```

```python
sto.drop(['crime_counts_job', 'totalpop_job', 'Race0','crime_counts_race', 'totalpop_race',
       'Edu0','crime_counts_edu', 'totalpop_edu', 'Or0','crime_counts_origin', 'totalpop_origin', 'Sal0','totalpop_sal', 'Sal15'], axis = 1, inplace = True)
```

```python
sto.shape
```

```python
sto.head()
```

```python
sto.rename(columns = {'Job0':"NTA_code"},  inplace = True)
sto.head()
```

```python
sto.set_index(keys='NTA_code', drop = True, inplace = True)
sto.head()
```

```python
for col in sto.columns[:-1]:
    sto[col] = sto[col]/sto.COUNTS
    
    
sto.head()
```

```python
sto.shape
```

```python
nta_Df =pd.DataFrame(sto.index)
nta_Df
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=5)
principalComponents = pca.fit_transform(sto)
principalDf = pd.DataFrame(data = principalComponents)
```

```python
principalDf = principalDf.merge(nta_Df, left_on = principalDf.index, right_on = nta_Df.index)
principalDf
```

```python
principalDf.drop('key_0',axis=1, inplace =True)
```

```python
principalDf.head()
```

```python
principalDf.columns = ['PCA%d'%p for p in range(0,len(principalDf.columns))]
principalDf.rename(columns = {'PCA5':'NTA_Code'}, inplace = True)
principalDf.head()
```

```python
principalDf.columns[:-1]
```

```python
lms = smf.ols(formula='NTA_Code ~ '+'+'.join(principalDf.columns[:-1]), data = principalDf).fit()
print(lms.summary())
```
