```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%%time
data311 = pd.read_csv('311_Service_Requests_from_2010_to_Present.csv', parse_dates = ['Created Date','Closed Date','Due Date'], low_memory = False)
data311.shape
```

```python
data311.shape
```

```python
data311.columns = data311.columns.str.replace(" ","_")
```

```python
data311.columns = data311.columns.str.lower()
```

```python
data311.info()
```

```python
data311.drop(['open_data_channel_type','park_facility_name', 'park_borough', 'vehicle_type',
       'taxi_company_borough', 'taxi_pick_up_location', 'bridge_highway_name',
       'bridge_highway_direction', 'road_ramp', 'bridge_highway_segment'], axis = 1, inplace = True)
```

```python
strings = ['agency', 'agency_name',
       'complaint_type', 'descriptor', 'location_type', 'incident_address', 'street_name', 'cross_street_1', 'cross_street_2',
       'intersection_street_1', 'intersection_street_2', 'address_type',
       'city', 'landmark', 'facility_type', 'status','resolution_description','borough']

for x in strings:
  data311[x] = data311[x].astype('str')
  data311[x] = data311[x].str.lower()
```

```python
data311.info()
```

```python
data311_copy = data311.copy()
```

```python
data311_copy.info()
```

```python
data311_copy.incident_zip = pd.to_numeric(data311_copy.incident_zip, errors = 'coerce')
```

```python
data311_copy.info()
```

```python
data311_copy = data311_copy.loc[(data311_copy.incident_zip>=10000) & (data311_copy.incident_zip<11500)]
```

```python
data311_cc = data311_copy.copy()
```

```python
data311_cc.incident_zip = data311_cc.incident_zip.astype('int')
```

```python
data311_cc.info()
```
