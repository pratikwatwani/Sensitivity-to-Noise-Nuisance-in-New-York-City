```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('acs_select_housing_08to12_ntas.xlsx - ACS_12_5YR_DP04.csv')
df.head()
```

```python
df.set_index(keys = 'Unnamed: 0', inplace = True)
df.index.rename('Parameters', inplace = True)
df.head()
```

```python
df.dropna(how = 'all', inplace = True)
df.head()
```

```python
df.info()
```

```python
df = df.astype('str')

for col in df.columns[1:]:
    df[col] = df[col].str.replace(',','')
    df[col] = df[col].str.replace('nan','0')
    df[col] = df[col].str.replace('-','')
```

```python
for column in df.columns[1:]:
    df[column] = pd.to_numeric(df[column])
```

```python
df.head()
```

```python
df.isna().values.all()
```

```python
df.notnull().values.all()
```

```python
df.columns = df.columns.str.replace(" ","_")
```

```python
df.to_csv('/wingrdp/homedirs/ppw220/housing.csv')
```
