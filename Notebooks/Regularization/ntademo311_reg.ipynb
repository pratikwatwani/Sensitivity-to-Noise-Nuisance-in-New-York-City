```python
temp_df_pivot.reset_index(level = 0, inplace = True)
temp_df_pivot.head()
```

```python
for col in temp_df_pivot.columns[1:-13]:
    temp_df_pivot[col] = temp_df_pivot[col].astype('float')
```

```python
x = pd.factorize(temp_df_pivot.NTA_code)
temp_df_pivot.NTA_code = x[0]
```

```python
for alpha in range(1,1000):
    Lasso=linear_model.Lasso(fit_intercept=True, alpha=alpha, max_iter=100000)
    Lasso.fit(temp_df_pivot.iloc[:,1:-13],temp_df_pivot.NTA_code)
```

```python
print(Lasso.sparse_coef_)
```

```python
Lasso.coef_
```

```python
for alpha in range(1,1000):
    Ridge=linear_model.Ridge(fit_intercept=True, alpha=alpha, max_iter=10000)
    Ridge.fit(temp_df_pivot.iloc[:,1:-13],temp_df_pivot.NTA_code)

Ridge.coef_
```
